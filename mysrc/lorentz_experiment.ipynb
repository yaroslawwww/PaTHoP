```python
from Lorentz import *
import matplotlib.pyplot as plt
from TimeSeries import *
r_values = [0.3, 1.8, 3.7, 10, 16, 24.06, 28, 30, 100, 150]

fig, axes = plt.subplots(2, 5, figsize=(20, 10), subplot_kw={'projection': '3d'})

for i, r in enumerate(r_values):
    x, y, z = Lorentz().generate(dt=0.01, steps=1000000, r=r)

    row = i // 5
    col = i % 5

    ax = axes[row, col]
    ax.scatter(x, y, z, lw=0.5)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.view_init(10, 100)
    ax.set_title(f'Решение уравнения лоренца при  r = {r}')
plt.tight_layout()
# plt.savefig('lor_exp.pdf',dpi=300,bbox_inches='tight')

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from TimeSeries import *

r_values = [24, 30, 24.74, 28, 150, 200]
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # Создаем сетку 2x3 графиков

for i, r in enumerate(r_values):
    ts = TimeSeries("Lorentz", size=100000, r=r, dt=0.01)

    row = i // 3
    col = i % 3
    axes[row, col].plot(np.arange(len(ts.values))[-100:], ts.values[-1000::10])
    axes[row, col].set_title(f'Ряд Лоренца при r = {r}')
    axes[row, col].set_xlabel('Время')
    axes[row, col].set_ylabel('Значение')

plt.tight_layout()  # Автоматически настраивает расположение графиков, чтобы избежать перекрытия
# plt.savefig('lorenz_plots.png',dpi=300,bbox_inches='tight')  # Сохраняем графики в файл
plt.show()
```
